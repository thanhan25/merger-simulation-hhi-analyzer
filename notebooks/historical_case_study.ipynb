```python
# Cell 1: Setup and Data Ingestion
import pandas as pd
from merger_sim.simulation import simulate_merger
from merger_sim.plotting import plot_merger_impact

# Historical Case: The blocked AT&T and T-Mobile Merger (2011)
data = {
    "Firm": ["AT&T", "Verizon", "T-Mobile", "Sprint", "Others"],
    "Market_Share": [29.0, 31.0, 11.0, 16.0, 13.0],
}
df = pd.DataFrame(data)
df
```

```python
# Cell 2: Execute Simulation Engine
acquirer = "AT&T"
target = "T-Mobile"

results = simulate_merger(df, "Firm", "Market_Share", acquirer, target)

print(f"Pre-Merger HHI: {results['hhi_pre']:.2f}")
print(f"Post-Merger HHI: {results['hhi_post']:.2f}")
print(f"HHI Delta: +{results['delta']:.2f}")
print(f"Regulatory Decision: {results['regulatory_risk']}")
```

```python
# Cell 3: Visual Proof
plot_merger_impact(results)
```
